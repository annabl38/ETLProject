```python
import os
import pandas as pd

cpu_path=('intel_CPUs.csv')
cpu_df=pd.read_csv(cpu_path)
cpu_df.head()
```

```python
cpu_df.dtypes
```

```python
cpu_df
```

```python
cpu_df=cpu_df.dropna(subset=['Recommended_Customer_Price'])
cpu_df =cpu_df[~cpu_df['Recommended_Customer_Price'].str.contains('-')]
cpu_df
```

```python
cpu_df=cpu_df.dropna(subset=['Launch_Date'])
```

```python
cpu_df['Recommended_Customer_Price']=cpu_df['Recommended_Customer_Price'].map(lambda x: str(x)[1:])
cpu_df
```

```python
cpu_df.dtypes
```

```python
cpu_df['Recommended_Customer_Price']=cpu_df['Recommended_Customer_Price'].str.replace(',','')
cpu_df
```

```python
cpu_df['Recommended_Customer_Price']=cpu_df['Recommended_Customer_Price'].astype('float')
```

```python
cpu_df.head()
```

```python
cpu_df=cpu_df[['Product_Collection','Vertical_Segment','Processor_Number','Launch_Date','Recommended_Customer_Price','nb_of_Cores','nb_of_Threads','Max_Memory_Size','Max_Memory_Bandwidth' ]]
```

```python
cpu_df=cpu_df.replace("04'16","Q4'16")
cpu_df=cpu_df.replace("Q1 '15","Q1'15")
```

```python
cpu_df['Quarter'] = cpu_df["Launch_Date"].map(lambda x: str(x)[:2])
cpu_df.head()
```

```python
cpu_df['Year'] = cpu_df["Launch_Date"].map(lambda x: str(x)[3:])
```

```python
cpu_df['clean_date']=cpu_df['Year']+"'"+cpu_df['Quarter']
```

```python
cpu_df.head()
```

```python
merge_cpu=cpu_df[['clean_date','Recommended_Customer_Price']].copy()
```

```python
merge_cpu=merge_cpu.groupby('clean_date').mean()
merge_cpu
```

```python
merge_cpu.to_csv('CPU_Final.csv')
```
