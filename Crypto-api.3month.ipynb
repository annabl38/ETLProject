```python
#BitCoin started trading in 2009 and is traded as BTC
btc_url = 'https://rest.coinapi.io/v1/ohlcv/BTC/USD/history?period_id=3MTH&time_start=2009-01-01T00:00:00'
btc_response = requests.get(btc_url, headers=headers).json()
btc_api_df = pd.read_json(json.dumps(btc_response))
btc_api_df.head()
```

```python
#Ripple started trading in 2012, for real in 2013 and is trading as XRP
xrp_url = 'https://rest.coinapi.io/v1/ohlcv/XRP/USD/history?period_id=3MTH&time_start=2013-01-01T00:00:00'
xrp_response = requests.get(xrp_url, headers=headers).json()
xrp_api_df = pd.read_json(json.dumps(xrp_response))
xrp_api_df.head()
```
