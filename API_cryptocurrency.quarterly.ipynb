# Bitcoin (BTC)

```python
# #BitCoin started trading in 2009 and is traded as BTC
# btc_url = 'https://rest.coinapi.io/v1/ohlcv/BTC/USD/history?period_id=3MTH&time_start=2009-01-01T00:00:00'
# btc_response = requests.get(btc_url, headers=headers).json()
# btc_api_df = pd.read_json(json.dumps(btc_response))

btc_url = 'https://rest.coinapi.io/v1/ohlcv/BTC/USD/history?period_id=3MTH&time_start=2009-01-01T00:00:00'
btc_response = requests.get(btc_url, headers=headers).json()
btc_api_df = pd.read_json(json.dumps(btc_response))
# btc_api_df.head()
```

```python
btc_api_df=btc_api_df.rename(columns={"trades_count":"BTC_trades", "price_close": "BTC_price_close", 
                                      "volume_traded":"BTC_trade_volume"})
btc_api_df.head()
```

```python
btc_api_df["CLEAN_TIME"] = btc_api_df["time_period_start"].map(lambda x: str(x)[:-21])

btc_api_df["CLEAN_TIME"] = btc_api_df["CLEAN_TIME"].str.replace('-01',"Q1")
btc_api_df["CLEAN_TIME"] = btc_api_df["CLEAN_TIME"].str.replace('-04',"Q2")
btc_api_df["CLEAN_TIME"] = btc_api_df["CLEAN_TIME"].str.replace('-07',"Q3")
btc_api_df["CLEAN_TIME"] = btc_api_df["CLEAN_TIME"].str.replace('-10',"Q4")

btc_api_df["CLEAN_TIME"] = btc_api_df["CLEAN_TIME"].map(lambda x: str(x)[2:])

btc_api_df["year"]=btc_api_df["CLEAN_TIME"]
btc_api_df["year"]=btc_api_df["year"].map(lambda x: str(x)[:-2])

btc_api_df["Quarter"]=btc_api_df["CLEAN_TIME"].map(lambda x: str(x)[2:])

btc_api_df["Year_quarter"]=(btc_api_df["year"])+"'"+(btc_api_df["Quarter"])
btc_api_quarter=btc_api_df[["BTC_price_close", "Year_quarter", "BTC_trades","BTC_trade_volume"]]
btc_api_quarter.head()
```

# formula for testing coins:

```python
def coin (coin):
    url = (f'https://rest.coinapi.io/v1/ohlcv/{coin}/USD/history?period_id=3MTH&time_start=2010-01-01T00:00:00')
    response = requests.get(url, headers=headers).json()
    api_df = pd.read_json(json.dumps(response))

    coin_volume_label = coin + "_trade_volume"
    coin_trade_label = coin + "_trade_count"
    coin_price_label = coin + "_price_close"
    api_df=api_df.rename(columns={"trades_count":coin_trade_label, "price_close": coin_price_label, 
                                          "volume_traded":coin_volume_label})

    api_df["CLEAN_TIME"] = api_df["time_period_start"].map(lambda x: str(x)[:-21])

    api_df["CLEAN_TIME"] = api_df["CLEAN_TIME"].str.replace('-01',"Q1")
    api_df["CLEAN_TIME"] = api_df["CLEAN_TIME"].str.replace('-04',"Q2")
    api_df["CLEAN_TIME"] = api_df["CLEAN_TIME"].str.replace('-07',"Q3")
    api_df["CLEAN_TIME"] = api_df["CLEAN_TIME"].str.replace('-10',"Q4")

    api_df["CLEAN_TIME"] = api_df["CLEAN_TIME"].map(lambda x: str(x)[2:])

    api_df["year"]=api_df["CLEAN_TIME"]
    api_df["year"]=api_df["year"].map(lambda x: str(x)[:-2])

    api_df["Quarter"]=api_df["CLEAN_TIME"].map(lambda x: str(x)[2:])

    api_df["Year_quarter"]=(api_df["year"])+"'"+(api_df["Quarter"])
    api_quarter=api_df[[coin_price_label, "Year_quarter", coin_trade_label,coin_volume_label]]
    return api_quarter
    
```
