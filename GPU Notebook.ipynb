```python
import pandas as pd
```

```python
csv_file = "All_GPUs.csv"
csv_df = pd.read_csv(csv_file)
csv_df
```

```python
csv_df.columns
```

```python
GPU_trimmed_col = csv_df[['Architecture', 'Core_Speed', 'Manufacturer', 'Memory', 'Memory_Bandwidth', 'Memory_Bus', 
                          'Memory_Speed',
       'Memory_Type', 'Name', 'Pixel_Rate', 'Process', 'Release_Date', 'Release_Price']]
GPU_trimmed_col.head()
```

```python
GPU_prices = GPU_trimmed_col[GPU_trimmed_col['Release_Price'].notnull()]
GPU_prices
```

```python
GPU_prices["clean_date"] = GPU_prices['Release_Date'].map(lambda x: str(x)[4:])
GPU_prices
```

```python
GPU_prices["Release_Price"]=GPU_prices["Release_Price"].map(lambda x: str(x)[1:])
GPU_prices["Release_Price"]=GPU_prices["Release_Price"].astype(float)
GPU_prices.head()
```

```python
#Change to Q1
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Jan','Q1')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Feb','Q1')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Mar','Q1')

#Change to Q2
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Apr','Q2')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('May','Q2')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Jun','Q2')

#Change to Q3
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Jul','Q3')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Aug','Q3')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Sep','Q3')

#Change to Q4
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Oct','Q4')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Nov','Q4')
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('Dec','Q4')

GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('-20',"'")
GPU_prices
```

```python
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('-',"")
GPU_prices["clean_date"]=GPU_prices["clean_date"].str.replace('\t',"")
GPU_prices
```

```python
GPU_with_quarter = GPU_prices.copy()
```

```python
GPU_with_quarter_trim = GPU_with_quarter[["Release_Price", "clean_date"]]
GPU_with_quarter_trim.dtypes
```

```python

GPU_with_quarter_trim
```

```python
GPU_with_quarter_trim['Quarter'] = GPU_with_quarter_trim["clean_date"].map(lambda x: str(x)[:2])
GPU_with_quarter_trim['Year'] = GPU_with_quarter_trim["clean_date"].map(lambda x: str(x)[3:])
```

```python
GPU_with_quarter_trim.head()
```

```python
GPU_with_quarter_trim["Year_Quarter"]=GPU_with_quarter_trim["Year"] + "'" + GPU_with_quarter_trim["Quarter"]
```

```python
pre_grouped_data = GPU_with_quarter_trim[['Release_Price', 'Year_Quarter']]
pre_grouped_data["Year_Quarter"]=pre_grouped_data["Year_Quarter"].str.replace(' ','')
pre_grouped_data.head()
```

```python
GPU_grouped = pre_grouped_data.groupby("Year_Quarter").mean()
GPU_grouped
```

```python
GPU_grouped = GPU_grouped[:-1]
GPU_grouped
```

```python
GPU_grouped=GPU_grouped.rename(columns={"Release_Price":"Average_Q_Price"})
GPU_grouped
```

```python
GPU_Final = GPU_grouped.copy()
```

```python
GPU_Final.to_csv("GPU_Final.csv")
```
